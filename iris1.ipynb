```python
from sklearn import datasets
```

```python
df = datasets.load_iris()
```

```python
df
```

```python
df.feature_names
```

```python
df.target_names
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df.data
y = df.target
```

```python
x
```

```python
y
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

```python
from sklearn import tree
text = tree.export_text(model)
```

```python
text
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize =(50,50))
a = tree.plot_tree(model, feature_names = df.feature_names, class_names=df.target_names, filled=True)
```

```python
import graphviz
fig.savefig('abc.png')
```

```python
dot_data = tree.export_graphviz(model, out_file = None , feature_names=  df.feature_names, class_names = df.target_names, filled =True)
```

```python
graph= graphviz.Source(dot_data, format='png')
graph
```
