```python
x = df.drop('salary_more_then_100k', axis= 'columns')
y = df['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
df['company'] = le1.fit_transform(x['company'])
df['job'] = le1.fit_transform(x['job'])
df['degree'] = le1.fit_transform(x['degree'])
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(df.drop('salary_more_then_100k',axis='columns'),df.salary_more_then_100k)
```

```python
model.predict([[2,2,1]])
```

```python
feature_names = ['company','job','degree']
```

```python
import numpy as np
class_names = np.array(['greater','lesser'])
```
